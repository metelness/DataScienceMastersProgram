````python

def read_raw_email(email_path):
    detector = UniversalDetector()
    
    try:
        with open(email_path) as f:
            original_msg = f.read()
    except UnicodeDecodeError:
        detector.reset()
        with open(email_path, 'rb') as f:
            for line in f.readlines():
                detector.feed(line)
                if detector.done:
                    break
        detector.close()
        encoding = detector.result['encoding']
        with open(email_path, encoding=encoding) as f:
            original_msg = f.read()
            
    return original_msg 

email_path=current_dir.joinpath('examples/html/message-0.txt')
read_raw_email(email_path)
````

````python
def make_spark_df():
    records = []
    mydict = {'id' : None,'username' : None,'original_msg' : None,}
    for root, dirs, files in os.walk(enron_data_dir):
        
        for file_path in files:
            current_path = str(Path(root).joinpath(file_path))
            user_name = os.listdir(enron_data_dir)[0]
            #print('USERNAME')
            #print(user_name)
            mydict['username'] = user_name
            ## Current path is now the file path to the current email.  
            ## Use this path to read the following information
            ## id (The relative path of the email message)
            
            id_ = current_path
            #enron_data_dir
            #print('ID')
            #print(id_)
            mydict['id'] = id_
            
            ## original_msg
            original_msg = read_raw_email(current_path)
            #print('Original Message')
            #print(original_msg)
            mydict['original_msg']=original_msg
            #time.sleep(5)
            records.append(mydict)
            
    df = pd.DataFrame(records)
    print(df.head(5))

    # load spark df
    sdf = spark.createDataFrame(df)
    
    #print(type(sdf))
    #print(sdf)
    return sdf
        
df=make_spark_df()       
df.printSchema()
````

````python
def parse_html_payload(payload):
    """
    This function uses Beautiful Soup to read HTML data
    and return the text.  If the payload is plain text, then
    Beautiful Soup will return the original content
    """
    soup = BeautifulSoup(payload, 'html.parser')
    return str(soup.get_text()).encode('utf-8').decode('utf-8')

def parse_email(original_msg):
    result = {'text' : None,'payload' : None}
    parsed_msg = parse_html_payload(original_msg)
    msg = Parser(policy=default).parsestr(original_msg)
    ## TODO: Use Python's email library to read the payload and the headers
    ## https://docs.python.org/3/library/email.examples.html
    #  Or for parsing headers in a string (this is an uncommon operation), use:
    headers = Parser(policy=default).parsestr(parsed_msg)
    #print(headers)
    #time.sleep(5)
    result['text'] = headers.get_content()
    #print(result)
    result['payload'] = headers.get_payload()
    #print(result)
    
    tuple_result = tuple([str(result.get(column, None)) for column in columns])
    return ParsedEmail(*tuple_result)
````
